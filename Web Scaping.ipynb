```python
from autoscraper import AutoScraper

url = 'https://www.amazon.in/s?k=iphone&ref=nb_sb_noss'

# We can add one or multiple candidates here.
# You can also put urls here to retrieve urls.
wanted_list = ["₹51,990","Apple iPhone 11 (128GB) - Purple"," 64,401"]

scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar("https://www.amazon.in/s?k=iphone&ref=nb_sb_noss",grouped=True)
```

```python
scraper.set_rule_aliases({'rule_6222': 'Stars', 'rule_ic2u': 'Title','rule_dr88':'Price'})
scraper.keep_rules(['rule_6222', 'rule_ic2u','rule_dr88'])
scraper.save('Amazon-proudct-search')
```

```python
scraper.load('Amazon-proudct-search')
```

```python
result=scraper.get_result_similar("https://www.amazon.in/s?k=iphone&ref=nb_sb_noss",group_by_alias=True)
```
